```python
def trim_nodes_by_degree(G,d):
 """ returns a copy of G without 
     the nodes with a degree less than d """
 Gt = G.copy()
 dn = nx.degree(Gt)
 for n in Gt.nodes():
  if dn[n] <= d:
   Gt.remove_node(n)
 return Gt
```

```python
Trm=trim_nodes_by_degree(G,300)
```

```python
def trim_edges_by_weight(G,w):
    """ returns a copy of G without
        edges of weight less than w """
    Gt = G.copy()
    for e in Gt.edges(data=True):
        if e[2]['weight'] < w:
            try:
                Gt.remove_edge(e[0],e[1])
            except:
                print e,e[0:2], "Problem"
                print "Unexpected error:", sys.exc_info()
    return Gt
```

```python
def delete_solitaries(G):
    Gd = G.copy()
    nds = nx.degree(Gd)
    sols=[]
    for (n,d) in nds.items():
        if (d==0):
            sols.append(n)
    Gd.remove_nodes_from(sols)
    return Gd
```
